```python
class Account:
    # Class variable
    history = []

    # Constructor
    def __init__(self, balance):
        self.__balance = balance
        Account.history.append(self.__balance)

    # Getter
    def get_balance(self):
        return self.__balance

    # Deposit
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            Account.history.append(self.__balance)
        else:
            print("Deposit amount must be greater than zero.")

    # Withdraw
    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.__balance:
                self.__balance -= amount
                Account.history.append(self.__balance)
            else:
                print("Insufficient balance.")
        else:
            print("Withdrawal amount must be greater than zero.")


# Create object
acc = Account(2000)

print("Initial Balance:", acc.get_balance())

acc.deposit(1000)
print("After Deposit:", acc.get_balance())

acc.withdraw(500)
print("After Withdrawal:", acc.get_balance())

print("Balance History:", Account.history)
```

```python
from abc import ABC,abstractmethod
class Shape(ABC):
    #constructor
    def __init__(self,radius=0,length=0,width=0):
        self.__radius=radius
        self.__length=length
        self.__width=width
    #getter
    def get_radius(self):
        return self.__radius
    def get_length(self):
        return self.__length
    def get_width(self):
        return self.__width
    # abstract method
    @abstractmethod
    def area(self):
        pass
    @abstractmethod  
    def perimeter(self):
        pass
#child class1
class Circle(Shape):
    def __init__(self, radius):
        super().__init__(radius=radius)
    def area(self):
        area=3.14*self.get_radius()*self.get_radius()
        print("The area of circle :",area)
    def perimeter(self):
        perimeter=2*3.14*self.get_radius()
        print(" The perimeter of circle :",perimeter)
    def display_all_details(self):
        print("Radius:", self.get_radius())
        self.area()
        self.perimeter()

    
#child class2
class Rectangle(Shape):
    def __init__(self, length, width):
        super().__init__(length=length, width=width)
    def area(self):
        area=self.get_length()*self.get_width()
        print(" the area of rectangle ",area)
    def perimeter(self):
        perimeter=2*(self.get_length()+self.get_width())
        print(" the perimeter of rectangle", perimeter)
    def display_all_details(self):
        print("Length:", self.get_length())
        print("Width :", self.get_width())
        self.area()
        self.perimeter()
shape1=Circle(2)
shape2=Rectangle(2,3)
print("Circle Details")
shape1.display_all_details()
print("-"*40)
print("Rectangle Details")
shape2.display_all_details()
```

```python
from abc import ABC,abstractmethod
#abstractmethod
@abstractmethod
class Shape3D(ABC):
    def volume(self):
        pass
    def surface_area(self):
        pass
#childclass1
class Cube(Shape3D):
    def __init__(self,side):
        self.side=side
    def volume(self):
        print("The volume of cube is :",self.side **3)
    def surface_area(self):
        print("The surface area cube is :",6*self.side ** 2)
#chidclass2
class Sphere(Shape3D):
    def __init__(self,radius):
        self.radius=radius
    def volume(self):
        print("The volume of sphere is :",(4/3)*3.14*self.radius**3)
    def surface_area(self):
        print("The surface area sphere is :",4*3.14*self.radius**2)
#chidclass3
class Cylinder(Shape3D):
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def volume(self):
        print("Volume of Cylinder:", 3.14 * self.radius ** 2 * self.height)

    def surface_area(self):
        print("Surface Area of Cylinder:",
              2 * 3.14 * self.radius * (self.radius + self.height)) 
# Create objects
cube = Cube(4)
sphere = Sphere(3)
cylinder = Cylinder(2, 5)

print("Cube")
cube.volume()
cube.surface_area()

print("-" * 40)

print("Sphere")
sphere.volume()
sphere.surface_area()

print("-" * 40)

print("Cylinder")
cylinder.volume()
cylinder.surface_area()
```

```python
from abc import ABC, abstractmethod

# Abstract class
class Database(ABC):

    @abstractmethod
    def connect(self):
        pass

    @abstractmethod
    def query(self, sql):
        pass


# Child class 1
class MySQLDatabase(Database):

    def connect(self):
        print("Connected to MySQL Database.")

    def query(self, sql):
        print(f"Executing MySQL Query: {sql}")


# Child class 2
class MongoDatabase(Database):

    def connect(self):
        print("Connected to MongoDB Database.")

    def query(self, sql):
        print(f"Executing MongoDB Query: {sql}")


# Create objects
mysql = MySQLDatabase()
mongo = MongoDatabase()

print("MySQL Database")
mysql.connect()
mysql.query("SELECT * FROM students;")

print("-" * 40)

print("MongoDB Database")
mongo.connect()
mongo.query("db.students.find()")
```

```python
from abc import ABC, abstractmethod

# Abstract class
class Transport(ABC):

    @abstractmethod
    def move(self):
        pass

    @abstractmethod
    def fuel_type(self):
        pass


# Child class 1
class Car(Transport):

    def move(self):
        print("Car moves on roads.")

    def fuel_type(self):
        print("Fuel Type: Petrol")


# Child class 2
class Train(Transport):

    def move(self):
        print("Train moves on railway tracks.")

    def fuel_type(self):
        print("Fuel Type: Electricity")


# Child class 3
class Airplane(Transport):

    def move(self):
        print("Airplane flies in the sky.")

    def fuel_type(self):
        print("Fuel Type: Aviation Fuel")


# Create objects
car = Car()
train = Train()
airplane = Airplane()

print("Car")
car.move()
car.fuel_type()

print("-" * 40)

print("Train")
train.move()
train.fuel_type()

print("-" * 40)

print("Airplane")
airplane.move()
airplane.fuel_type()
```

```python
from abc import ABC, abstractmethod

# Abstract class
class FileHandler(ABC):

    @abstractmethod
    def read(self):
        pass

    @abstractmethod
    def write(self, data):
        pass


# Child class 1
class TextFileHandler(FileHandler):

    def read(self):
        print("Reading data from a text file.")

    def write(self, data):
        print(f"Writing '{data}' to a text file.")


# Child class 2
class CSVFileHandler(FileHandler):

    def read(self):
        print("Reading data from a CSV file.")

    def write(self, data):
        print(f"Writing '{data}' to a CSV file.")


# Create objects
text_file = TextFileHandler()
csv_file = CSVFileHandler()

print("Text File Handler")
text_file.read()
text_file.write("Hello World")

print("-" * 40)

print("CSV File Handler")
csv_file.read()
csv_file.write("Name,Age\nJohn,25")
```
