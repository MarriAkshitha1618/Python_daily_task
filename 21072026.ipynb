```python
#hard2
class BankAccount:
    total_accounts = 0
    history = []

    def __init__(self, balance):
        if balance < 1000:
            self.__balance = 1000
            print("Minimum balance is 1000. Balance set to 1000.")
        else:
            self.__balance = balance

        if self.__balance > 100000:
            bonus = self.__balance * 0.05
            self.__balance += bonus
            print(f"5% bonus added: {bonus}")

        BankAccount.total_accounts += 1
        BankAccount.history.append(
            f"Account {BankAccount.total_accounts} created with balance {self.__balance}"
        )

    def get_balance(self):
        return self.__balance
    def set_balance(self, balance):
        if balance < 1000:
            print("Balance cannot be less than minimum balance (1000).")
            return

        if balance > 100000:
            bonus = balance * 0.05
            balance += bonus
            print(f"5% bonus added: {bonus}")

        old_balance = self.__balance
        self.__balance = balance

        BankAccount.history.append(
            f"Balance updated from {old_balance} to {self.__balance}"
        )

    @classmethod
    def show_total_accounts(cls):
        return cls.total_accounts

    @classmethod
    def show_history(cls):
        print("\nTransaction History:")
        for record in cls.history:
            print(record)


acc1 = BankAccount(5000)
acc2 = BankAccount(150000)

print("Account 1 Balance:", acc1.get_balance())
print("Account 2 Balance:", acc2.get_balance())

acc1.set_balance(800)
acc1.set_balance(120000)

print("Updated Account 1 Balance:", acc1.get_balance())

print("Total Accounts:", BankAccount.show_total_accounts())

BankAccount.show_history()
```

```python
#hard3
class Student:
    total_students = 0

    def __init__(self, marks):
        self.__marks = marks
        self.status = "Passed"

        if any(mark < 40 for mark in marks):
            self.status = "Failed"

        Student.total_students += 1

   
    def get_marks(self):
        return self.__marks

    def get_status(self):
        return self.status

    
    def set_marks(self, marks):
       
        if len(marks) != 5:
            print("Marks should contain exactly 5 subjects.")
            return

        for mark in marks:
            if mark < 0 or mark > 100:
                print("Each mark must be between 0 and 100.")
                return

        old_avg = sum(self.__marks) / len(self.__marks)
        new_avg = sum(marks) / len(marks)

       
        if new_avg < old_avg:
            print("Average cannot decrease.")
            return

        self.__marks = marks

       
        if any(mark < 40 for mark in marks):
            self.status = "Failed"
        else:
            self.status = "Passed"

        print("Marks updated successfully.")


s1 = Student([70, 80, 75, 90, 85])

print("Marks:", s1.get_marks())
print("Status:", s1.get_status())

s1.set_marks([80, 85, 90, 95, 88])
print("Marks:", s1.get_marks())
print("Status:", s1.get_status())


s1.set_marks([60, 65, 70, 75, 80])


s1.set_marks([90, 95, 38, 96, 98])
print("Marks:", s1.get_marks())
print("Status:", s1.get_status())

print("Total Students:", Student.total_students)
```

```python
#hard4
class Product:
    total_revenue = 0

    def __init__(self, price):
        self.__price = price
        self.__original_price = price
        self.__discount = 0

        if price > 5000:
            self.__discount = 15

        Product.total_revenue += price

   
    def get_price(self):
        return self.__price

 
    def get_discount(self):
        return self.__discount

    
    def set_price(self, price):
        
        if price < self.__original_price * 0.80:
            print("Price cannot be less than 80% of the original price.")
            return

        self.__price = price

        if price > 5000:
            self.__discount = 15
        else:
            self.__discount = 0

       
        Product.total_revenue += price

        print("Price updated successfully.")

    
    def set_discount(self, discount):
        self.__discount = discount

  
    @classmethod
    def get_total_revenue(cls):
        return cls.total_revenue


p1 = Product(4000)
p2 = Product(6000)

print("P1 Price:", p1.get_price())
print("P1 Discount:", p1.get_discount(), "%")

print("P2 Price:", p2.get_price())
print("P2 Discount:", p2.get_discount(), "%")


p1.set_price(3500)


p1.set_price(3000)

p2.set_price(7000)

print("P1 Price:", p1.get_price())
print("P1 Discount:", p1.get_discount(), "%")

print("P2 Price:", p2.get_price())
print("P2 Discount:", p2.get_discount(), "%")

print("Total Revenue:", Product.get_total_revenue())
```

```python
#hard5
class User:
    senior_citizen = 0      

    def __init__(self, password, age):
        self.__password = ""
        self.__age = 0
        self.is_senior = False

        self.set_password(password)
        self.set_age(age)

    
    def get_password(self):
        return self.__password

    
    def get_age(self):
        return self.__age

    
    def set_password(self, password):
        if len(password) < 8:
            print("Password must be at least 8 characters.")
            return False

        if not any(ch.isdigit() for ch in password):
            print("Password must contain at least one digit.")
            return False

        self.__password = password
        return True

    
    def set_age(self, age):
        if age < self.__age:
            print("Age cannot be decreased.")
            return False

        self.__age = age

        if age >= 60 and not self.is_senior:
            self.is_senior = True
            User.senior_citizen += 1

        return True

    
    def update(self, password, age):
        if self.set_password(password) and self.set_age(age):
            print("User updated successfully.")
        else:
            print("Update failed.")



u1 = User("Pass1234", 45)

print("Age:", u1.get_age())
print("Password:", u1.get_password())

u1.update("NewPass99", 60)

print("Age:", u1.get_age())
print("Password:", u1.get_password())


u1.update("abc", 65)


u1.update("Strong123", 50)

print("Senior Citizens:", User.senior_citizen)
```
