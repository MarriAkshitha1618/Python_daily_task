# Shallow Copy

```python
import copy
list_original = [1, 2, [3, 4], 5]
list_duplicate=copy.copy(list_original)
print(list_original)
print(list_duplicate)


```

```python
list_duplicate[2][0]=99
print(list_duplicate)
print(id(list_original))
print(id(list_duplicate))
```

# Deep Copy

```python
import copy
list_original = [1, 2, [3, 4], 5]
list_duplicate=copy.deepcopy(list_original)
print(list_original)
print(list_duplicate)
```

```python
import copy
list_original = [1, 2, [3, 4], 5]
list_duplicate=copy.deepcopy(list_original)
print(list_original)
list_duplicate[2][0]=99
print(list_duplicate)
```

# demonstrate shallow copy using copy.copy() on a nested list.

```python
list1=[[1,2,3,4],[5,6,7,8]]
list2=copy.copy(list1)
print(list1)
print(list2)
print(id(list1))
print(id(list2))
print(id(list1[1][0]))
print(id(list2[1][0]))
```

# to demonstrate deep copy using copy.deepcopy() on the same nested list

```python
list1=[[1,2,3,4],[5,6,7,8]]
list2=copy.deepcopy(list1)
print(list1)
print(list2)
print(id(list1))
print(id(list2))
print(id(list1[1][0]))
print(id(list2[1][0]))
```

```python
import copy
def detect_copy(original,copied):
    copied[2][0]=100
    if original[2][0]==100:
        print("shallow")
    else:
        print("deep")
original=[1,2,[3,4]]
shallow=copy.copy(original)
detect_copy(original,shallow)
original=[1,2,[3,4]]
deep=copy.deepcopy(original)
detect_copy(original,deep)
```

```python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount 
            print(f"₹{amount:.2f} deposited successfully.")
            print(f"Current Balance: ₹{self.balance:.2f}")
        else:
            print("Invalid deposit amount.")
    def withdraw(self, amount):
        if amount > self.balance:
            print("Transaction Denied: Insufficient balance.")
            print(f"Available Balance: ₹{self.balance:.2f}")
        elif amount <= 0:
            print("Please enter a valid amount to withdraw.")
        else:
            self.balance -= amount  
            print(f"₹{amount:.2f} withdrawn successfully.")
            print(f"Remaining Balance: ₹{self.balance:.2f}")

    def display_balance(self):
        print(f"Account Holder: {self.account_holder} | Balance: ₹{self.balance:.2f}")
my_account = BankAccount("Amit Sharma", 10000)
my_account.display_balance()
my_account.deposit(5000)
my_account.withdraw(4000)
```

```python
class Car:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def increase_speed(self, value):
        if value < 0:
            print("Invalid input: Cannot increase speed by a negative value.")
            return
        self.speed += value 
        if self.speed > 200:
            print("Warning: Speed cap reached! Capping speed at 200 km/h.")
            self.speed = 200 
        print(f"The {self.brand}'s speed is now {self.speed} km/h.")
    def display_speed(self):
        print(f"Car: {self.brand} | Current Speed: {self.speed} km/h")
my_car = Car("Tata Safari", 120)
my_car.display_speed()
my_car.increase_speed(50)
my_car.increase_speed(45)
```

```python
class MovieTicket:
    def __init__(self,movie_name,seats):
        self.movie_name=movie_name
        self.seats=seats
    def book_ticket(self, number):
        if number <=0:
            print(f"Invalid input")
            return
        if number <= self.seats:
            self.seats -= number
            print(f" Success booking {number} tickets for {self.movie_name}")
            print(f"Remaining Seats:: {self.seats}")
        else:
            print("Booking Denied:")
            print(f"Available seats {self.seats}")
    def display_details(self):
        print(f"Movie : {self.movie_name} | Available seats: {self.seats}")
show=MovieTicket("Inception",10)
show.display_details()
show.book_ticket(4)
show.book_ticket(8)
show.book_ticket(6)
        
```

```python
class Product:
    def __init__(self, product_name, quantity):
        self.product_name = product_name
        self.quantity = quantity

    def sell(self, sell_quantity):
        if sell_quantity <= 0:
            print("Invalid quantity.")
            return

        if sell_quantity <= self.quantity:
            self.quantity -= sell_quantity   # Corrected
            print(f"Success! Sold {sell_quantity} units of '{self.product_name}'.")
        else:
            print("Not enough stock.")

    def display_stock(self):
        print(f"{self.product_name} - Stock: {self.quantity}")


item = Product("Wireless Mouse", 5)

item.display_stock()
item.sell(3)
item.display_stock()
item.sell(4)
item.sell(2)
item.display_stock()

```

```python
class SchoolFees:
    def __init__(self,student_name,total_fee,paid_fee):
        self.student_name = student_name
        self.total_fee = total_fee
        self.paid_fee = paid_fee
    def pay_fee(self, amount):
        if self.paid_fee + amount <= self.total_fee:
            self.paid_fee += amount
            print("Fee paid successfully.")
            print("Paid Fee:", self.paid_fee)
            print("Remaining Fee:", self.total_fee - self.paid_fee)
        else:
            print("Payment exceeds the total fee.")
name = input("Enter student name: ")
total_fee = float(input("Enter total fee: "))
paid_fee = float(input("Enter paid fee: "))
amount = float(input("Enter amount to pay: "))

student = SchoolFees(name, total_fee, paid_fee)
student.pay_fee(amount)
```

```python
class Bus:
    def __init__(self, bus_name, total_seats):
        self.bus_name = bus_name
        self.total_seats = total_seats
        self.booked_seats = 0

    def reserve(self, number):
        if self.booked_seats + number <= self.total_seats:
            self.booked_seats += number
            print("Reservation Successful")
            print("Remaining Seats:", self.total_seats - self.booked_seats)
        else:
            print("Seats not available")


b = Bus("Express", 40)

b.reserve(10)
b.reserve(20)
b.reserve(15)
```

```python
class ShoppingCart:
    def __init__(self, customer_name, wallet_balance):
        self.customer_name = customer_name
        self.cart_total = 0
        self.wallet_balance = wallet_balance

    # Add item
    def add_item(self, price):
        self.cart_total += price
        print("Item Added")
        print("Cart Total:", self.cart_total)

    # Remove item
    def remove_item(self, price):
        if self.cart_total - price >= 0:
            self.cart_total -= price
            print("Item Removed")
        else:
            print("Cart total cannot be negative")

    # Checkout
    def checkout(self):
        if self.wallet_balance >= self.cart_total:
            self.wallet_balance -= self.cart_total
            print("Payment Successful")
            self.cart_total = 0
        else:
            print("Insufficient Balance")

        print("Wallet Balance:", self.wallet_balance)
        print("Cart Total:", self.cart_total)


c = ShoppingCart("Akshitha", 5000)

c.add_item(1000)
c.add_item(500)
c.remove_item(200)
c.checkout()
```
