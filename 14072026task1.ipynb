```python
value=["sumit","Sumit","Sumit123","SuMiT","Sumit@123","1234"]
result=[]
for i in value:
    if value.islower():
        result.append("lower")
    elif value.isupper():
        result.append("upper")
    elif value.isalpha():
        result.append("alpha")
    elif value.isalnum():
        result.append("alphanum")
    elif value.isdigit():
        result.append("digit")
    else:
        result.append("special Char")
print(result)
```

```python
n = int(input("Enter a limit (n): "))
num = 2
print(f"Prime numbers between 2 and {n}:")

while num <= n:
    divisor = 2  
    is_prime = True
    
   
    while divisor * divisor <= num:
        if num % divisor == 0:
            is_prime = False
            break
        divisor += 1  
    if is_prime:
        print(num, end=" ")
        
    num += 1  
print()
```

```python

num = int(input("Enter a number: "))


factorial = 1


if num < 0:
    print("Factorial does not exist for negative numbers.")
elif num == 0:
    print("The factorial of 0 is 1.")
else:
    
    for i in range(1, num + 1):
        factorial *= i 
        
    print(f"The factorial of {num} is {factorial}.")
```
