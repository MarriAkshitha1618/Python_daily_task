```python
num=5
fact=1
temp = num
if num<0:
    print("Fact doesnot exit")
elif num==0:
    print("fact of 0 is 1")
else:
    while temp>0:
        fact *= temp
        temp -= 1
    print(f"The factorial of {num} is {fact}")
```

```python

while True:
  
    print("\n--- CALCULATOR MENU ---")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")
    
   
    choice = input("Select an option (1-5): ")
    
    if choice == '5':
        print("\nExiting calculator. Goodbye!")
        break  
    elif choice in ('1', '2', '3', '4'):
        print() 
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        if choice == '1':
            result = num1 + num2
            print(f"Result: {num1} + {num2} = {result}")
            
        elif choice == '2':
            result = num1 - num2
            print(f"Result: {num1} - {num2} = {result}")
            
        elif choice == '3':
            result = num1 * num2
            print(f"Result: {num1} x {num2} = {result}")
            
        elif choice == '4':
            if num2 == 0:
                print("Error: Division by zero is not allowed!")
            else:
                result = num1 / num2
                print(f"Result: {num1} / {num2} = {result}")
    else:
        print("\nInvalid choice! Please select a valid option from 1 to 5.")
```

```python
print("--- Leap Year Checker ---")
print("Enter a year to check if it's a leap year, or type 'stop' to exit.\n")
while True:
    user_input = input("Enter a year (or 'stop'): ").strip()
    if user_input.lower() == 'stop':
        print("Program stopped. Goodbye!")
        break
    if user_input.isdigit():
        year = int(user_input)
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            print(f"-> {year} is a leap year.\n")
        else:
            print(f"-> {year} is NOT a leap year.\n")
    else:
        print("Invalid input! Please enter a valid year number or type 'stop'.\n")
```
