#Functions with Default Parameters

```python
def greet():
    return "Hello, Welcome to Python!"
greet()
```

```python
def show_date():
    return "17/07/2026 and 4: 22"
show_date()
```

```python
def print_square():
    for i in range(1,11):
        print(i*i)
print_square()
```

```python
def motivation():
    return f" Honsetly is the best policy"
motivation()
```

```python
def line():
    for i in range(1,51):
        print("-",end="")
line()
```

```python
def greet_user(name="Akshitha"):
    return  f" My name is {name}, Good Morning"
greet_user()
```

```python
def power(base,exp=2):
    return base**exp
print(power(5))
```

```python
def student_info(name,age, city):
   
    return f"Name: {name}, Age: {age}, City: {city}"
student_info(name="Akshitha", age=18, city="Delhi")
```

```python
def add_numbers(a, b=0, c=0):
    result=a+b+c
    return result
add_numbers(10,20,20)
```

```python
def repeat_message(msg, times=3):
    for i in range(times):
        print(msg)
repeat_message("hi")
repeat_message("Loop",2)
```

```python
def calculate_discount(price, discount=10):
    discount_amount=price*(discount/100)
    final_price=price-discount_amount
    return final_price
calculate_discount(1000,10)
```

```python
def introduce(name="Unknown", profession="Student"):
    return f" My name is {name}, i am a {profession} "
introduce()
```

```python
def calculate(a, b, operation):
    if operation == "add":
        return a + b
    elif operation == "subtract":
       return a - b
    elif operation == "multiply":
        return a * b
    elif operation == "division":
        return a / b 
calculate(a=10,b=20,operation="add")
```
